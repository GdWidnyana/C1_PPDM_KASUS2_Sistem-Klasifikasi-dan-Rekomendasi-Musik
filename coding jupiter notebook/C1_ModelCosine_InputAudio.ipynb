```python
import numpy as np
import matplotlib.pyplot as plt
spike_cols = [col for col in data.columns if 'mean' in col]
corr = data[spike_cols].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(16, 11))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(0, 25, as_cmap=True, s=90, l=45, n=5)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.title('Correlation Heatmap Mean Variables', fontsize=20)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn import preprocessing


#---
import pandas as pd
from sklearn import preprocessing
from sklearn.metrics.pairwise import cosine_similarity

# Load the CSV file
df = pd.read_csv('data_CS.csv', index_col='filename')

# Extract labels and drop irrelevant columns
labels = df[['label']]
df = df.drop(columns=['length', 'label'])

# Scale the data
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)

# Print the scaled data type
print('Scaled data type:', type(df_scaled))
```

```python
# Compute cosine similarity
similarity = cosine_similarity(df_scaled)
print("Similarity shape:", similarity.shape)

sim_df_labels = pd.DataFrame(similarity)
sim_df_names = sim_df_labels.set_index(labels.index)
sim_df_names.columns = labels.index

sim_df_names.head()
```

```python
import joblib
# Simpan scaler ke file
joblib.dump(df_scaled, 'scaler.joblib')
```
